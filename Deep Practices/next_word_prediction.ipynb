```python
# importing libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
input_sequence = []

for sentence in data.split('\n'):
  tokenize_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenize_sentence)):
     input_sequence.append(tokenize_sentence[:i+1])
```

```python
max_len = max([len(x) for x in input_sequence])
max_len
```

```python
padded_input_sequence = np.array(pad_sequences(input_sequence,maxlen=max_len,padding='pre'))
```

```python
padded_input_sequence
```

```python
X = padded_input_sequence[:,:-1]
```

```python
y = padded_input_sequence[:,-1]
```

```python
X.shape , y.shape
```

```python
y = to_categorical(y,num_classes=len(tokenizer.word_index)+1)
y.shape
```

```python
model = Sequential()

model.add(Embedding(input_dim=299, output_dim=100, input_length=79))

model.add(LSTM(200, return_sequences=True))
model.add(LSTM(200))
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.build(input_shape=(None, 79))

model.summary()
```

```python
model.fit(X,y,epochs=50)
```

```python
import time
text = "a random variable"

for i in range(15):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
